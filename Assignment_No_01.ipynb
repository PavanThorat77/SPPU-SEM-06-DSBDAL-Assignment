```python
#5.2 Performing appropriate type conversions
df['Postcode'] = df['Postcode'].astype('int')
df['Bedroom2']=df['Bedroom2'].astype('int')
df['Bathroom']=df['Bathroom'].astype('int')
df['Propertycount']=df['Propertycount'].astype('int')
df.info()
```

```python
#5.3 Data Normalization:
#Check for count of missing values:
df.isnull().sum()
```

```python
#5.3.1 Normalize Car's Null Values:
df['Car'] = df['Car'].fillna(df['Car'].median())
df['Car'].isnull().sum()
```

```python
#5.3.2 Normalize BuildingArea's Null Values:
df['BuildingArea'] = df['Car'].fillna(df['Car'].mean())
df['BuildingArea'].isnull().sum()
```

```python
#5.3.3 Normalize YearBuilt's Null Values:
df['YearBuilt'] = df['YearBuilt'].fillna(df['YearBuilt'].mode()[0])
df['YearBuilt'].isnull().sum()
```

```python
#5.3.4 Normalize YearBuilt's Null Values:
df['CouncilArea'] = df['CouncilArea'].fillna(df['CouncilArea'].mode()[0])
df['CouncilArea'].isnull().sum()
```

```python
#5.4 Printing number of null values
df.isnull().sum()
```

```python
#6.1 Forming categories in Type column
df['Type'] = df['Type'].astype('category')
df.dtypes
```

```python
df['Type']
```

```python
#6.2 Converting categories into quantitative variables
from sklearn import preprocessing
label_encoder = preprocessing.LabelEncoder()
df['Type'] = label_encoder.fit_transform(df['Type'])
df['Type'].unique()
```
